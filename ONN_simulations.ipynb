```python
def generate_planar_dataset(N=400, noise_ratio=0.01):
    points = 2 * (np.random.rand(2*N).reshape((N,2))) - 1
    labels = np.zeros((N,2))
    for i, point in enumerate(points):
        if 0.4 <= np.linalg.norm(point) + noise_ratio * np.random.randn() <= 0.8:
            labels[i][0] = 1
        else:
            labels[i][1] = 1
    return points, labels
```

```python
def one_hot_to_value(y):
    total = 0
    for i, yi in enumerate(y):
        total += i * yi
    return total / (len(y)-1)
```

```python
X, Y = generate_planar_dataset()
labels = np.array([one_hot_to_value(yi) for yi in Y]).flatten()
plt.figure(figsize=(6,6))
plt.scatter((X.T)[0, :], (X.T)[1, :], c=labels, cmap=plt.cm.Spectral)
plt.colorbar()
plt.show()
```

```python
def plot_boundary(X, Y, model):
    x_min, x_max = X[0, :].min() - 1, X[0, :].max() + 1
    y_min, y_max = X[1, :].min() - 1, X[1, :].max() + 1
    h = 0.1
    # Generate a grid of points with distance h between them
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    # Predict the function value for the whole grid
    Z = np.array([one_hot_to_value(model.forward_pass(np.array([x,y,0,0], dtype=np.complex64))) for x, y in zip(xx.flatten(), yy.flatten())])
    Z = Z.reshape(xx.shape)
    # Plot the contour and training examples
        
    plt.figure(figsize=(6,6))
    plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral)
    
    labels = np.array([one_hot_to_value(yi) for yi in Y]).flatten()

    plt.ylabel('x2')
    plt.xlabel('x1')
    
    plt.scatter((X.T)[0, :], (X.T)[1, :], c=labels, cmap=plt.cm.Spectral)
    plt.colorbar()

```

```python
def msc_loss(yhat, y):
    loss = 1/2 * (yhat - y)**2
    loss [2:] = 0
    return loss

def d_msc_loss(yhat, y):
    d_loss = yhat - y
    d_loss[2:] = 0
    return d_loss
```

```python
def one_hot_cross_entropy(y_pred, y_true):
    return np.sum(-1 * y_true * np.log(y_pred))

def d_one_hot_cross_entropy(y_pred, y_true):
    return -1 * y_true / y_pred
```

```python
N = 4

model = neu.Sequential([
    neu.ClementsLayer(N),
    neu.Activation(neu.nonlinearities.Abs(N)),
    neu.ClementsLayer(N),
    neu.Activation(neu.nonlinearities.Abs(N)),
    neu.ClementsLayer(N),
    neu.Activation(neu.nonlinearities.Abs(N)),
    neu.ClementsLayer(N),
    neu.Activation(neu.nonlinearities.Abs(N)),
    neu.ClementsLayer(N),
    neu.Activation(neu.nonlinearities.Abs(N)),
    neu.ClementsLayer(N),
    neu.Activation(neu.nonlinearities.AbsSquared(N)),
    neu.Activation(neu.nonlinearities.Mask(N, mask=[1,1,1e-3,1e-3])),
    neu.Activation(neu.nonlinearities.SoftMax(N))
])

P0 = 10
X_formatted = np.pad(X, (0,N-2), mode="constant")
for i, x in enumerate(X_formatted):
    X_formatted[i][2] = np.sqrt(P0 - np.sum(x**2))
Y_formatted = np.pad(Y, (0,N-2), mode="constant")

optimizer = neu.AdjointOptimizer(model, one_hot_cross_entropy, d_one_hot_cross_entropy, learning_rate=0.01)
losses = optimizer.fit(X_formatted, Y_formatted, iterations=100)
```

```python
plot_boundary(X, Y, model)
```
